**Why Logistic Regression for Credit Card Fraud:**

Logistic Regression is good when output is only 2 options → like fraud (1) or
not fraud (0).

It gives result in probability (like 0.8 means 80% chance fraud).

It’s fast, easy to train, and works fine when data is not too complex.

Create two new datasets: one with only non-fraud transactions (Class = 0) and one with only fraud transactions (Class = 1)

```python
non_fraud = data[data.Class == 0] # now we have splitted the data to new variables of dataset in which one is containing the non-fraud
fraud = data[data.Class == 1] # it is dataset with a fruad
```

```python
print(non_fraud.shape) # it is printing the data
print(fraud.shape)
```

```python

non_fraud.Amount.describe() # with this we are printing the Amount of the dataset and describing it also
```

```python
fraud.Amount.describe() # same description for the fraud dataset
```

 Group the data by 'Class' (0 = non-fraud, 1 = fraud) and calculate the mean of all other columns for each group

```python
data.groupby('Class').mean()  # Group the data by 'Class' (0 = non-fraud, 1 = fraud) and calculate the mean of all other columns for each group
```

```python
legit_sample = non_fraud.sample(n=492)  # Randomly select 492 rows from the non-fraud dataset
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0) # Combine the sampled non-fraud transactions and all fraud transactions into a new dataset
```

```python
new_dataset.head()
```

```python
new_dataset.tail() #printing the last some rows of  new_dataset
```

```python
new_dataset['Class'].value_counts() #counting the values of the new_dataset
```

```python
new_dataset.groupby('Class').mean()  # Group the new dataset by 'Class' (0 = non-fraud, 1 = fraud) and calculate the average of all other columns for each class
```

Separate the data into input features (`X`) by removing the 'Class' column and target labels (`Y`) using the 'Class' column to predict fraud or non-fraud


```python
X = new_dataset.drop(columns='Class', axis=1)  # Features: all columns except 'Class'
Y = new_dataset['Class']                        # Target: the 'Class' column (0 = non-fraud, 1 = fraud)
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training and testing sets: 80% for training and 20% for testing, keeping the class distribution the same (stratify=Y) and using a fixed random order (random_state=2)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) #we are splitting the dataset into Training Dataset of X and Y and testing Dataset of X & Y
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**First, we make a logistic regression model.

Then, we train it with training data (X_train and Y_train).

Next, we predict the labels for training data.

We check training accuracy by comparing predictions with real labels.

After that, we predict labels for test data (X_test).

Finally, we check test accuracy by comparing test predictions with real labels.**

```python
model = LogisticRegression() # we are calling the model which we imported in teh very first step
```

```python

model.fit(X_train, Y_train) # now wwe are fitting the data in the model for training the model;
```

```python

X_train_prediction = model.predict(X_train) # at last we are getting the model predicions over our model traing with the dataset
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) # now with the use of the training data accuracy lib we are getting the score of teh model means the acuracy score via its model predictions
```

```python
print('Accuracy on Training data : ', training_data_accuracy) #printing the mdoel accurcay
```

```python
X_test_prediction = model.predict(X_test) # at last we are getting the model predicions over our model traing with the testing dataset
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) # again we are calcualting the accuracy score
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy) #printing the score
```
